```python
# 60.	Compress string (aaabb → a3b2)
s=input("enter a string")
i = 0
result = ""
while i < len(s):
    count =1
    while i < len(s) - 1 and s[i] == s[i+1]:
        count+=1
        i += 1
    result = result + s[i] + str(count)
    i += 1
print("compressed string: ", result)
```
